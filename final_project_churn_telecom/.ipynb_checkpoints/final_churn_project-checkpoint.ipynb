# Прогнозирование оттока клиентов

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

В ходе данной работы мы проделаем следующие шаги:
1. Предобработка данных:
    - изучим сырые данные
    - составим список и уточним смысл части данных
    - выявим и устраним проблемы (пропуски, дубликаты, аномалии, ненужная информация и прочее)
    - проверим некоторые аналитические гипотезы, возможно, получим некоторые инсайты
    - определимся с типом машинного обучения, согласуем метрики (AUC-ROC не менее 0,85)
2. Машинное обучение:
    - подготовим данные к машинному обучению (объединение, выборки, кодирование, масштабирование)
    - обучим несколько моделей, чтобы найти наилучшую кросс-валидацией
    - проверим качество модели на тестовых данных, напишем вывод
3. Подведение итогов:
    - оформим полученные результы в формате отчета

## Подготовка данных

### Работа с признаками

```python
# проверим столбец TotalCharges
data_contract['TotalCharges'].value_counts()
```

Видим, что есть 11 значений с пустой строкой с пробелом. Это уже пропуск, который не был заметен ранее. Скорее всего эти клиенты вообще ничего не платили. Посмотрим:

```python
data_contract.loc[data_contract['TotalCharges'] == ' ']
```

Гипотеза не оправдалась. Так как датасеты актульны на 1 февраля 2020, а клиенты из списка именно в этот день заключили договор, то скорее всего поэтому инфа могла не попасть в последний столбец. На мой взгляд, будет достаточно перенести туда инфу из MonthlyCharges вместо удаления данных или взятия среднего, т.е. общий расход у них будет равен расходу за первый месяц.

```python
data_contract.loc[data_contract['TotalCharges'] == ' ', 'TotalCharges'] = \
    data_contract.loc[data_contract['TotalCharges'] == ' ', 'MonthlyCharges']
```

```python
# проверим одно их полученных значений
data_contract.iloc[488]
```

```python
# приведем весь столбец к типу float
data_contract['TotalCharges'] = pd.to_numeric(data_contract['TotalCharges'], errors='coerce')
data_contract.info()
```

Сразу же в этом датасете можем выделить новый столбец с целевым признаком *is_lost*, где 0 — клиент действующий, а 1 — клиент ушел.

```python
# заполняем сначала нулями
data_contract['is_lost'] = 0
# указываем ушедших клиентов
data_contract['is_lost'] = data_contract['is_lost'].where(data_contract['EndDate'] == 'No', 1)
data_contract.head()
```

```python
# посмотрит теперь на другой датафрейм
data_personal['SeniorCitizen'].value_counts()
```

Видим, что других значений нет, поэтому здесь всё ок.

### Объединение датасетов
Для дальнейшей работы нам нужно будет объединить все датасеты. Объединять будем по *customerID*.

```python
# создадим копию датафрейм
all_data = data_contract.copy()
# объединяем данные договоров и персональные данные
all_data = all_data.merge(data_personal, on='customerID', how='left')
```

```python
# объединяем с данными по интернету
all_data = all_data.merge(data_internet, on='customerID', how='left')
```

```python
# объединяем с данными по телефонии
all_data = all_data.merge(data_phone, on='customerID', how='left')
all_data.head()
```

```python
# проверим наличие пропусков в каждом признаке
all_data.isna().sum()
```

Как и ожидалось, появились пропуски значений из-за того, что не все клиенты пользуются и интернетом, и телефонией. В данном случае пропуски — это отсутствие услуг, значит, достаточно заменить на *No*.

```python
all_data = all_data.fillna('No')
all_data.isna().sum().sum()
```

Пропусков больше нет.

Перед тем, как приступить к исследовательскому анализу данных, имеет смысл избавиться от ненужных признаков, а также добавить новые, если в этом появится необходимость. Однозначно можно будет избавиться от следующих признаков:
- **customerID** (это лишь техническая информация, по причине ID пользователи точно не принимают решения)
- **BeginDate** (дата начала договора тоже не оказывает влияния, так как телеком-услуги в общем и целом не являются строго сезонными услугами)
- **EndDate** (по той же причине, что и выше)

Но при этом даты можно использовать перед удалением для того, чтобы рассчитать длительность пользования услуг. Будет лучше посчитать в днях, так как могут быть клиенты, которые даже месяца не попользовались. Так как целевой признак мы уже выделили ранее, то с этими столбцами можем делать что угодно, например, заполнить для действующих клиентов дату «окончания» датой последней сверки (2020-02-01).

```python
all_data.loc[all_data['EndDate'] == 'No', 'EndDate'] = '2020-02-01'
all_data.head()
```

```python
# приводим даты к типу datetime
all_data['BeginDate'] = pd.to_datetime(all_data['BeginDate'], format='%Y-%m-%d')
all_data['EndDate'] = pd.to_datetime(all_data['EndDate'], format='%Y-%m-%d')

# создаем новый столбец duration с разницей между 2 датами
all_data['duration'] = (all_data['EndDate'] - all_data['BeginDate']) / np.timedelta64(1,'D')

all_data.head()
```

```python
# удаляем ненужные столбцы
all_data = all_data.drop(['BeginDate', 'EndDate', 'customerID'], axis=1)
all_data.head()
```

Теперь мы можем переходить к следующей части, где подготовим данные для обучения модели и займемся самим обучением.

# Обучение моделей

## Обработка данных

Перед тем, как приступать к обучению моделей, нужно подготовить полученный датасет.

```python
# обозначим единым рандом
RANDOM_STATE = 190623
```

### Деление на выборки

```python
# разделим признаки
features = all_data.drop(['is_lost'], axis=1)
target = all_data['is_lost']

# обозначим категориальные признаки
cat_features = ['Type', 'PaperlessBilling', 'PaymentMethod', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 
                'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                'StreamingTV', 'StreamingMovies', 'MultipleLines']
```

```python
# разбиваем данные на обучающую и тестовую выборки в соотношении 75:25
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE, stratify=target
)

# проверим верность пропорций
print(features_train.shape)
print(target_train.shape)

print(features_test.shape)
print(target_test.shape)
```

### Кодирование данных

```python
# запишем функцию для кодирования
def one_hot_encoding(sample):
    sample = sample.reset_index(drop=True)
    enc = OneHotEncoder(drop='first')
    encoder_df = pd.DataFrame(enc.fit_transform(sample[cat_features]).toarray())
    return sample.join(encoder_df).drop(cat_features, axis=1)
```

```python
# закодируем все выборки с признаками
features_train = one_hot_encoding(features_train)
features_test = one_hot_encoding(features_test)
features_train.head()
```

Получили закодированные в числовые значения выборки, которые уже можно применять в работе. Кодировать целевой признак не нужно. Однако, количественные переменные расходов и длительности договора сильно выбиваются, поэтому данные нужно будет отмасштабировать.

### Масштабирование данных

Используем StandardScaler для масштабирования, при этом обучим его только на тренировочной выборке.

```python
# приведем числовые имена новых столбцов к строкам
features_train.columns = features_train.columns.map(str)
features_test.columns = features_test.columns.map(str)
```

```python
# создаем объект
scaler = StandardScaler()

# обучаем стандартизатор
scaler.fit(features_train)

# трансформируем
features_train = pd.DataFrame(scaler.transform(features_train))
features_test = pd.DataFrame(scaler.transform(features_test))
features_train.head()
```

После масштабирования уже можно переходить к обучению моделей.

## Обучение моделей

Для начала попробуем добиться нужного результата при помощи логичестической регрессии и случайного леса для классификации.

```python
# построим модель логистической регрессии и кросс-валидацией получим метрики
model_log_res = LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', class_weight='balanced')
print(cross_val_score(model_log_res, features_train, target_train, cv=3, scoring='roc_auc'))
```

Как мы видим, качество метрики не дотягивает до нужного нам. Попробуем случайный лес с подбором гиперпараметров.

```python
# Создание модели
model_forest = RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1)

# Поиск лучших параметров с помощью GridSearchCV
grid_forest = GridSearchCV(model_forest, param_grid={'max_depth': [5,10,12], 'n_estimators': [50,100,150]}, 
                           scoring='roc_auc')
grid_forest.fit(features_train, target_train)

# Вывод лучших параметров и метрики
print(f'Лучший оценщик: {grid_forest.best_estimator_}')
print(f'Лучшая метрика: {grid_forest.best_score_}')
```

Видим, что метрика выросла, но её все еще недостаточно. Попробуем градиентный бустинг.

```python
# задаем гиперпараметры
params_lgb = [{
    'objective': ['binary'],
    'num_leaves': [31, 50],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [5, 10], 
    'n_estimators': [50, 100]
}]

# Создание модели
model_lgb = LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1)

# Поиск лучших параметров с помощью GridSearchCV
grid_lgb = GridSearchCV(model_lgb, params_lgb, scoring='roc_auc', cv=5)
grid_lgb.fit(features_train, target_train)

# Вывод лучших параметров и метрики
print(f'Лучший оценщик: {grid_lgb.best_params_}')
print(f'Лучшая метрика: {grid_lgb.best_score_}')
```

В итоге градиентный бустинг показал наилучшее качество (0.8547227248726873). Сейчас нам нужно проверить на тестовой выборке.

## Проверка на тестовой выборке

Сейчас мы посмотрим, как модель покажет себя на данных, которых еще не видела. Дополнительно выведем не только AUC-ROC, но и accuracy, чтобы оценить точность.

```python
# обучаем и проверяем
final_model = LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1, 
                             learning_rate=0.1, max_depth=10, n_estimators=100, num_leaves=31, 
                             objective='binary')
final_model.fit(features_train, target_train)
predictions = final_model.predict(features_test)

probabilities_valid = final_model.predict_proba(features_test)
probabilities_one_valid = probabilities_valid[:, 1]

print('Accuracy: {0}| AUC-ROC: {1}'.format(
    accuracy_score(target_test, predictions), roc_auc_score(target_test, probabilities_one_valid)
))
```

Видим, что на тестовой выборке AUC-ROC показал хорошее качество. Нет ни переобучения, ни недообучения. Построим ROC-кривую, чтобы визуализировать полученный в итоге результат.

```python
# получаем значения FPR, TPR и пороги
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_valid)

# построим график
plt.figure()

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr,tpr)

plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.show()
```

На графике хорошо видно, насколько прогноз модели дает более точные результаты по сравнению с полностью случайным ответом, т.е. где с 50% вероятностью может быть дан ответ о том, уйдет ли клиент или нет.
