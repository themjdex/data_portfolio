# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

## –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–ü–µ—Ä–µ–¥ —Ç–µ–º, –∫–∞–∫ –ø—Ä–∏—Å—Ç—É–ø–∞—Ç—å –∫ –æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π, –Ω—É–∂–Ω–æ –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç.

```python
# –æ–±–æ–∑–Ω–∞—á–∏–º –µ–¥–∏–Ω—ã–π —Ä–∞–Ω–¥–æ–º
RANDOM_STATE = 190623
```

### –î–µ–ª–µ–Ω–∏–µ –Ω–∞ –≤—ã–±–æ—Ä–∫–∏

```python
# —Ä–∞–∑–¥–µ–ª–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏
features = all_data.drop(['is_lost'], axis=1)
target = all_data['is_lost']

# –æ–±–æ–∑–Ω–∞—á–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
cat_features = ['Type', 'PaperlessBilling', 'PaymentMethod', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 
                'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                'StreamingTV', 'StreamingMovies', 'MultipleLines']
```

```python
# —Ä–∞–∑–±–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 75:25
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE, stratify=target
)

# –ø—Ä–æ–≤–µ—Ä–∏–º –≤–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–æ–ø–æ—Ä—Ü–∏–π
print(features_train.shape)
print(target_train.shape)

print(features_test.shape)
print(target_test.shape)
```

<div class="alert alert-success">
<b>üëç –£—Å–ø–µ—Ö:</b> –í—Å–µ –≤–µ—Ä–Ω–æ!
</div>

### –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

```python
# –∑–∞–∫–æ–¥–∏—Ä—É–µ–º —Ç—Ä–µ–π–Ω
features_train = features_train.reset_index(drop=True)
enc = OneHotEncoder(drop='first')
encoder_df = pd.DataFrame(enc.fit_transform(features_train[cat_features]).toarray())
features_train = features_train.join(encoder_df).drop(cat_features, axis=1)
features_train.head()
```

```python
# –∑–∞–∫–æ–¥–∏—Ä—É–µ–º —Ç–µ—Å—Ç
features_test = features_test.reset_index(drop=True)
encoder_df_test = pd.DataFrame(enc.transform(features_test[cat_features]).toarray())
features_test = features_test.join(encoder_df_test).drop(cat_features, axis=1)
features_test.head()
```

<div class="alert alert-danger">
<b>üòî –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å:</b> –ö–æ–¥–∏—Ä–æ–≤—â–∏–∫ –Ω—É–∂–Ω–æ –æ–±—É—á–∞—Ç—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –∏ –ø–æ—Ç–æ–º –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å –∏–º –æ–±–µ –≤—ã–±–æ—Ä–∫–∏
</div>

<div class="alert-info">
<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞ 3</b>
    
–ò–∑–º–µ–Ω–∏–ª.
</div>

–ü–æ–ª—É—á–∏–ª–∏ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –≤ —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ —É–∂–µ –º–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω—è—Ç—å –≤ —Ä–∞–±–æ—Ç–µ. –ö–æ–¥–∏—Ä–æ–≤–∞—Ç—å —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–µ –Ω—É–∂–Ω–æ. –û–¥–Ω–∞–∫–æ, –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Ä–∞—Å—Ö–æ–¥–æ–≤ –∏ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –¥–æ–≥–æ–≤–æ—Ä–∞ —Å–∏–ª—å–Ω–æ –≤—ã–±–∏–≤–∞—é—Ç—Å—è, –ø–æ—ç—Ç–æ–º—É –¥–∞–Ω–Ω—ã–µ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –æ—Ç–º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞—Ç—å.

### –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

–ò—Å–ø–æ–ª—å–∑—É–µ–º StandardScaler –¥–ª—è –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è, –ø—Ä–∏ —ç—Ç–æ–º –æ–±—É—á–∏–º –µ–≥–æ —Ç–æ–ª—å–∫–æ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ.

```python
# –ø—Ä–∏–≤–µ–¥–µ–º —á–∏—Å–ª–æ–≤—ã–µ –∏–º–µ–Ω–∞ –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ —Å—Ç—Ä–æ–∫–∞–º
features_train.columns = features_train.columns.map(str)
features_test.columns = features_test.columns.map(str)

# –∑–∞–¥–∞–¥–∏–º –∫–æ–ª-—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
numeric = ['MonthlyCharges', 'TotalCharges', 'duration']
```

<div class="alert alert-danger">
<b>üòî –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å:</b> –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞—Ç—å –Ω—É–∂–Ω–æ —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –∞ –Ω–µ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ
</div>

<div class="alert-info">
<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞ 3</b>
    
–ò–∑–º–µ–Ω–∏–ª.
</div>

```python
# —Å–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç
scaler = StandardScaler()

# –æ–±—É—á–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ç–æ—Ä
scaler.fit(features_train[numeric])

# —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º
scaler_train = pd.DataFrame(scaler.transform(features_train[numeric]))
features_train = pd.concat([scaler_train, features_train.drop(numeric, axis=1)], axis=1)
features_train = features_train.rename(columns={0: 'MonthlyCharges', 1: 'TotalCharges', 2: 'duration'})

scaler_test = pd.DataFrame(scaler.transform(features_test[numeric]))
features_test = pd.concat([scaler_test, features_test.drop(numeric, axis=1)], axis=1)
features_test = features_test.rename(columns={0: 'MonthlyCharges', 1: 'TotalCharges', 2: 'duration'})

features_train.head()
```

–ü–æ—Å–ª–µ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è —É–∂–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç—å –∫ –æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π.

## –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–î–ª—è –Ω–∞—á–∞–ª–∞ –ø–æ–ø—Ä–æ–±—É–µ–º –¥–æ–±–∏—Ç—å—Å—è –Ω—É–∂–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –ø—Ä–∏ –ø–æ–º–æ—â–∏ –ª–æ–≥–∏—á–µ—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.

```python
# –ø–æ—Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π –ø–æ–ª—É—á–∏–º –º–µ—Ç—Ä–∏–∫–∏
model_log_res = LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', class_weight='balanced')
print(cross_val_score(model_log_res, features_train, target_train, cv=3, scoring='roc_auc'))
```

–ö–∞–∫ –º—ã –≤–∏–¥–∏–º, –∫–∞—á–µ—Å—Ç–≤–æ –º–µ—Ç—Ä–∏–∫–∏ –Ω–µ –¥–æ—Ç—è–≥–∏–≤–∞–µ—Ç –¥–æ –Ω—É–∂–Ω–æ–≥–æ –Ω–∞–º. –ü–æ–ø—Ä–æ–±—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏
model_forest = RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1)

# –ü–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å –ø–æ–º–æ—â—å—é GridSearchCV
grid_forest = GridSearchCV(model_forest, param_grid={'max_depth': [5,10,12], 'n_estimators': [50,100,150]}, 
                           scoring='roc_auc')
grid_forest.fit(features_train, target_train)

# –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –º–µ—Ç—Ä–∏–∫–∏
print(f'–õ—É—á—à–∏–π –æ—Ü–µ–Ω—â–∏–∫: {grid_forest.best_estimator_}')
print(f'–õ—É—á—à–∞—è –º–µ—Ç—Ä–∏–∫–∞: {grid_forest.best_score_}')
```

–í–∏–¥–∏–º, —á—Ç–æ –º–µ—Ç—Ä–∏–∫–∞ –≤—ã—Ä–æ—Å–ª–∞, –Ω–æ –µ—ë –≤—Å–µ –µ—â–µ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ. –ü–æ–ø—Ä–æ–±—É–µ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π –±—É—Å—Ç–∏–Ω–≥.

```python
# –∑–∞–¥–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã
params_lgb = [{
    'objective': ['binary'],
    'num_leaves': [31, 50],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [5, 10], 
    'n_estimators': [50, 100]
}]

# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏
model_lgb = LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1)

# –ü–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å –ø–æ–º–æ—â—å—é GridSearchCV
grid_lgb = GridSearchCV(model_lgb, params_lgb, scoring='roc_auc', cv=5)
grid_lgb.fit(features_train, target_train)

# –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –º–µ—Ç—Ä–∏–∫–∏
print(f'–õ—É—á—à–∏–π –æ—Ü–µ–Ω—â–∏–∫: {grid_lgb.best_params_}')
print(f'–õ—É—á—à–∞—è –º–µ—Ç—Ä–∏–∫–∞: {grid_lgb.best_score_}')
```

–í –∏—Ç–æ–≥–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π –±—É—Å—Ç–∏–Ω–≥ –ø–æ–∫–∞–∑–∞–ª –Ω–∞–∏–ª—É—á—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ (0.8547227248726873). –°–µ–π—á–∞—Å –Ω–∞–º –Ω—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

–°–µ–π—á–∞—Å –º—ã –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∂–µ—Ç —Å–µ–±—è –Ω–∞ –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä—ã—Ö –µ—â–µ –Ω–µ –≤–∏–¥–µ–ª–∞. –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –≤—ã–≤–µ–¥–µ–º –Ω–µ —Ç–æ–ª—å–∫–æ AUC-ROC, –Ω–æ –∏ accuracy, —á—Ç–æ–±—ã –æ—Ü–µ–Ω–∏—Ç—å —Ç–æ—á–Ω–æ—Å—Ç—å.

```python
# –æ–±—É—á–∞–µ–º –∏ –ø—Ä–æ–≤–µ—Ä—è–µ–º
final_model = LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced', n_jobs=-1, 
                             learning_rate=0.1, max_depth=10, n_estimators=100, num_leaves=31, 
                             objective='binary')
final_model.fit(features_train, target_train)
predictions = final_model.predict(features_test)

probabilities_valid = final_model.predict_proba(features_test)
probabilities_one_valid = probabilities_valid[:, 1]

print('Accuracy: {0} | AUC-ROC: {1}'.format(
    accuracy_score(target_test, predictions), roc_auc_score(target_test, probabilities_one_valid)
))
```

–í–∏–¥–∏–º, —á—Ç–æ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ AUC-ROC –ø–æ–∫–∞–∑–∞–ª —Ö–æ—Ä–æ—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ. –ù–µ—Ç –Ω–∏ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è, –Ω–∏ –Ω–µ–¥–æ–æ–±—É—á–µ–Ω–∏—è. –ü–æ—Å—Ç—Ä–æ–∏–º ROC-–∫—Ä–∏–≤—É—é, —á—Ç–æ–±—ã –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –≤ –∏—Ç–æ–≥–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –î–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ 85%.

```python
# –ø–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è FPR, TPR –∏ –ø–æ—Ä–æ–≥–∏
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_valid)

# –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure()

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr,tpr)

plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-–∫—Ä–∏–≤–∞—è')
plt.show()
```
